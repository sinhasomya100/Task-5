```python
import pandas as pd

# Titanic dataset ka GitHub link
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Dataset ko load karo
df = pd.read_csv(url)

# Pehle 5 rows dekhne ke liye
df.head()
```

```python
# Dataset ka structure samajhne ke liye
df.info()
```

```python
# Step 2: Numerical columns ki summary statistics
df.describe()
```

```python
# Step 3: Missing values check karo
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 4.1: Survived passengers count plot
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.show()
```

```python
# Step 4.2: Passenger class vs survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked Location')
plt.xlabel('Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Count')
plt.show()
```

```python
# Check how many missing values exist in each column
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
df['Survived'].value_counts()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
sns.boxplot(data=df, x='Survived', y='Age')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Value Heatmap")
plt.show()
```

## Summary of Insights

- Female passengers had higher survival rates than males.  
- **First-class** passengers had better survival chances.  
- Passengers aged **20-40** were the majority onboard.  
- Passengers who paid higher fares had better survival rates.  
- The `'Cabin'` column had too many missing values and was dropped.
